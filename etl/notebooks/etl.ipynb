```python
def process(data, sheet_name):
    countries = (data[['Country Name', 'Country Code', 'Country WHO Region', 'Country Income Group']]
                 .drop_duplicates(subset=['Country Code']))
    countries['country'] = countries['Country Code'].map(to_concept_id)
    countries = countries.set_index('country')
    
    indicators = (data[['Indicator Name', 'Indicator Code', 'Indicator Unit']]
                  .drop_duplicates(subset=['Indicator Code']))
    indicators['concept'] = (indicators['Indicator Code'] + " " + sheet_name).map(to_concept_id)
    indicators = indicators.set_index('concept')
    
    dps = data[['Indicator Code', 'Indicator Unit', 'Country Code', 'Year', 'Value']].copy()
    dps['Country Code'] = dps['Country Code'].map(to_concept_id)
    dps_groups = dps.groupby(['Indicator Code'])
    for g, df in dps_groups:
        concept = to_concept_id(g + ' ' + sheet_name)
        dp = df[['Country Code', 'Year', 'Value']].copy()
        dp.columns = ['country', 'year', concept]
        dp = dp.dropna(subset=[concept])
        
        if not dp.empty:
            dp.to_csv(f'../../ddf--datapoints--{concept}--by--country--year.csv', index=False)
        
    return countries, indicators
```

```python
c1, i1 = process(data, "Constant_2010_NCU_per_capita")
```

```python
sheets = ['Million_NCU',
 'Million_current_US',
 'Million_current_PPP',
 'Current_NCU_per_capita',
 'Current_US_per_capita',
 'Current_PPP_per_capita',
 'Millon_constant_2010_NCU',
 'Million_constant_2010_US',
 'Million_constant_2010_PPP',
 'Constant_2010_NCU_per_capita',
 'Constant_2010_US_per_capita',
 'Constant_2010_PPP_per_capita',
 'Percentage_of_GDP',
 'Percentage_of_GGE']
```

```python
countries = []
indicators = []

for s in sheets:
    print(s)
    data = pd.read_excel(source_file, sheet_name=s)
    
    c1, i1 = process(data, s)
    
    countries.append(c1)
    indicators.append(i1)
```

```python
ctry = pd.concat(countries)
```

```python
ctry = ctry.reset_index().drop_duplicates(subset=['country'])
```
